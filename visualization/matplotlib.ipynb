Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Matplotlib Basics

```python
x = np.linspace(0,5,11)
x

y = x**3
y
```

```python
plt.plot(x,y)
plt.title('Cubic')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# plt creates a canvas and plot creates axis on it

plt.subplot(2,2,1)
plt.subplot(2,2,4)
plt.plot(y,x)
```

Object Oriented Way

```python
# for controlling canvas
fig = plt.figure(figsize=(20,8),dpi=100)    # dpi for more resulation
axis1 = fig.add_axes([0.1,0.1,1,1])       #  percentile of original canvas in the order from left from from bottom and last two length and width
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,0.3,0.3])       # last two are size, percentile of original canvas
axis2.plot(x,y)
```

Types of Plot

```python
plt.scatter(x,y)
```

```python
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)
```

```python
plt.boxplot(x)
```

Saving plot

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,.8,.8])
ax.plot(x,y)

plt.savefig('basicplot.png')
```
