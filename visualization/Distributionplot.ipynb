```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

We use distribution plots to visualize the distribution of quantitative data.


Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

* distplot/histplot
* jointplot
* pairplot
* rugplot

Jointplot visualizes the relationship between two variables along with their distributions, while pairplot shows pairwise relationships among multiple variables in a grid format for exploratory analysis

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()   # so size is not quantitive as it's values are limited. so it is qualititive
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 20,kde = True)      # https://youtu.be/DCgPRaIDYXA?si=drjrcqmDLaKqgzaK for kernel density estimation (the smooth line over histogram)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 20,kde = True)

```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'scatter')   # for comparing multiple plots
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'reg')
```

```python
sns.pairplot(df,hue='sex')   
```

```python
sns.pairplot(df,hue='size',palette='rainbow')     # hue catagorically differentiates based on size important!
```

```python
sns.rugplot(df['tip'])
```
